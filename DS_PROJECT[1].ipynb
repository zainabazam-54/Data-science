```python

import datetime as dt


# Clean column names (remove spaces & standardize names)
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Drop missing values
df.dropna(subset=["CustomerID"], inplace=True)

# Convert InvoiceDate to datetime format
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# Calculate Total Sales Amount
df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]

# Define the reference date (Assuming latest date in dataset)
reference_date = df["InvoiceDate"].max()

# Aggregate RFM metrics per CustomerID
rfm = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (reference_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",  # Frequency
    "TotalAmount": "sum"  # Monetary
}).reset_index()

# Rename columns
rfm.rename(columns={
    "InvoiceDate": "Recency",
    "InvoiceNo": "Frequency",
    "TotalAmount": "Monetary"
}, inplace=True)

# Assign RFM scores (1-5 scale)
rfm["R_Score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2, 1])
rfm["F_Score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["M_Score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])

# Compute RFM Segment
rfm["RFM_Score"] = rfm["R_Score"].astype(str) + rfm["F_Score"].astype(str) + rfm["M_Score"].astype(str)

# Display final RFM table
print(rfm.head())

# Save RFM results
rfm.to_csv("RFM_Segmentation.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Example data (replace this with your actual dataset)
X = np.random.rand(100, 3)  # 100 samples, 3 features

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Ensure X_scaled and feature_vector are defined
if 'X_scaled' not in locals():
    raise ValueError("X_scaled must be defined before running this code.")

# Ensure X_scaled has at least 3 columns for visualization
if X_scaled.shape[1] < 3:
    raise ValueError("X_scaled must have at least 3 columns for visualization.")

# Define feature names for visualization (replace with your actual feature names)
feature_vector = ['Feature 0', 'Feature 1', 'Feature 2']

# Parameters for the elbow method
cl = 50  # Maximum number of clusters to try
corte = 0.1  # Threshold for determining the best K

# Initialize variables
anterior = float('inf')  # Use infinity for a more readable approach
cost = []
K_best = cl

# Elbow Method to find the optimal number of clusters
for k in range(1, cl + 1):
    model = KMeans(
        n_clusters=k,
        init='k-means++',
        n_init=10,
        max_iter=300,
        tol=1e-4,
        random_state=101
    )
    model.fit(X_scaled)
    inertia = model.inertia_  # Sum of squared distances to nearest cluster center
    if (K_best == cl) and (anterior != 0) and (((anterior - inertia) / anterior) < corte):
        K_best = k - 1
    cost.append(inertia)
    anterior = inertia

# Elbow Method Plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, cl + 1), cost, marker='o', linestyle='-', color='red')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Cost)")
plt.title("Elbow Method for Optimal K")
plt.grid()
plt.show()

# Final K-Means Model with Optimal K
print(f"The best K suggested: {K_best}")
model = KMeans(n_clusters=K_best, init='k-means++', n_init=10, max_iter=300, tol=1e-4, random_state=101)
model.fit(X_scaled)

labels = model.labels_  # Get cluster labels

# Visualizing the clusters
fig, axes = plt.subplots(1, 2, figsize=(20, 5))

# Scatter plot for first two features
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels.astype(float), cmap='viridis', edgecolor='k')
axes[0].set_xlabel(feature_vector[0])
axes[0].set_ylabel(feature_vector[1])
axes[0].set_title("Clusters: Feature 0 vs Feature 1")

# Scatter plot for second and third features
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 2], c=labels.astype(float), cmap='viridis', edgecolor='k')
axes[1].set_xlabel(feature_vector[0])
axes[1].set_ylabel(feature_vector[2])
axes[1].set_title("Clusters: Feature 0 vs Feature 2")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples

# Example data (replace this with your actual dataset)
rfm_scaled = np.random.rand(100, 3)  # 100 samples, 3 features
feature_vector = ['Feature 0', 'Feature 1', 'Feature 2']  # Replace with your feature names
K_best = 10  # Replace with the optimal number of clusters from the Elbow Method

# Dictionary to store cluster metrics
cluster_results = {}

# Loop through different cluster sizes
for n_clusters in range(3, K_best + 1, 2):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 7))

    # Silhouette plot setup
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(rfm_scaled) + (n_clusters + 1) * 10])

    # K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, tol=1e-4, random_state=101)
    cluster_labels = kmeans.fit_predict(rfm_scaled)

    try:
        # Compute silhouette score
        silhouette_avg = silhouette_score(rfm_scaled, cluster_labels)

        # Store cluster centers and metrics
        cluster_results[n_clusters] = {
            'centers': kmeans.cluster_centers_,
            'silhouette_score': silhouette_avg,
            'labels': cluster_labels
        }

        # Compute silhouette samples
        sample_silhouette_values = silhouette_samples(rfm_scaled, cluster_labels)
        y_lower = 10
        for i in range(n_clusters):
            ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.viridis(float(i) / n_clusters)
            ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)

            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10  # Padding

        ax1.set_title("Silhouette Plot for Clusters")
        ax1.set_xlabel("Silhouette Coefficient Values")
        ax1.set_ylabel("Cluster Label")
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
        ax1.set_yticks([])
        ax1.set_xticks(np.linspace(-0.1, 1, 12))

        # Scatter plots for visualization
        centers = kmeans.cluster_centers_
        colors = cm.viridis(cluster_labels.astype(float) / n_clusters)

        # First scatter plot (Feature 1 vs Feature 0)
        x, y = 1, 0
        ax2.scatter(rfm_scaled[:, x], rfm_scaled[:, y], marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')
        ax2.scatter(centers[:, x], centers[:, y], marker='o', c="white", alpha=1, s=200, edgecolor='k')

        for i, c in enumerate(centers):
            ax2.scatter(c[x], c[y], marker=f'${i}$', alpha=1, s=50, edgecolor='k')

        ax2.set_title(f"{n_clusters} Clustered Data")
        ax2.set_xlabel(feature_vector[x])
        ax2.set_ylabel(feature_vector[y])

        # Second scatter plot (Feature 2 vs Feature 0)
        x = 2
        ax3.scatter(rfm_scaled[:, x], rfm_scaled[:, y], marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')
        ax3.scatter(centers[:, x], centers[:, y], marker='o', c="white", alpha=1, s=200, edgecolor='k')

        for i, c in enumerate(centers):
            ax3.scatter(c[x], c[y], marker=f'${i}$', alpha=1, s=50, edgecolor='k')

        ax3.set_title(f"Silhouette Score: {silhouette_avg:.2f}")
        ax3.set_xlabel(feature_vector[x])
        ax3.set_ylabel(feature_vector[y])

        plt.suptitle(f"Silhouette Analysis for KMeans Clustering with n_clusters = {n_clusters}",
                     fontsize=14, fontweight='bold')
        plt.show()

    except Exception as e:
        print(f"Error calculating silhouette score for {n_clusters} clusters: {e}")
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your cleaned dataset
# Create a binary matrix for transactions
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

# Convert quantities to binary values (1 for presence, 0 for absence)
def encode_units(x):
    return 1 if x >= 1 else 0

basket_sets = basket.applymap(encode_units)

# Apply FP-Growth algorithm
frequent_itemsets = fpgrowth(basket_sets, min_support=0.005, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)

# Display the rules
print("Generated Association Rules:\n", rules)

# Print lift and confidence values
print("\nLift Values:\n", rules["lift"])
print("\nConfidence Values:\n", rules["confidence"])

# Visualization: Lift vs Confidence Scatter Plot
if not rules.empty:
    plt.figure(figsize=(10, 6))
    plt.scatter(rules["confidence"], rules["lift"],
                s=rules["support"] * 1000,  # Scale point size
                c=rules["support"], cmap="viridis", alpha=0.7, edgecolors="k")

    plt.colorbar(label="Support")  # Add color legend
    plt.xlabel("Confidence")
    plt.ylabel("Lift")
    plt.title("Lift vs Confidence of Association Rules")
    plt.grid(True)
    plt.show()
else:
    print("No association rules found. Try lowering min_support or min_threshold.")
```
